```python
# Image classifier project
import tensorflow as tf
from tensorflow.keras import datasets, layers, models
from tensorflow.keras import layers, models, regularizers
from tensorflow.keras.preprocessing.image import ImageDataGenerator
import matplotlib.pyplot as plt

# 1. Collect and Preprocess Data
# CIFAR-10 is a perfect dataset for building a working model
(train_images, train_labels), (test_images, test_labels) = datasets.cifar10.load_data()

# Normalize pixel values to be between 0 and 1
train_images, test_images = train_images / 255.0, test_images / 255.0

class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# 2. Build the CNN Model (Project Implementation)
model = models.Sequential([
    # Convolutional layers to extract features from images
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),

    # Dense layers for classification
    layers.Flatten(),
    layers.Dense(64, activation='relu'),
    layers.Dense(10) # 10 output classes
])

# 3. Compile and Tune Hyperparameters
# We use the 'adam' optimizer as suggested in learning resources
model.compile(optimizer='adam',
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=['accuracy'])

# 4. Train the Model
# Target: Working model
print("Training the image classifier...")
history = model.fit(train_images, train_labels, epochs=10,
                    validation_data=(test_images, test_labels))

# 5. Evaluate the Working Model
test_loss, test_acc = model.evaluate(test_images,  test_labels, verbose=2)
print(f'\nFinished Training. Model Accuracy: {test_acc*100:.2f}%')

# Model Optimization
# 1. Data Augmentation
# Creates variations of images so the model generalizes better
datagen = ImageDataGenerator(
    rotation_range=15,
    width_shift_range=0.1,
    height_shift_range=0.1,
    horizontal_flip=True
)

# 2. Optimized CNN Architecture
model_opt = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(32, 32, 3)),
    layers.BatchNormalization(), # Batch Normalization
    layers.MaxPooling2D((2, 2)),

    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.BatchNormalization(),
    layers.MaxPooling2D((2, 2)),

    layers.Flatten(),
    layers.Dense(64, activation='relu', kernel_regularizer=regularizers.l2(0.001)), # Regularization
    layers.Dropout(0.5), # Dropout (Week 5 Requirement)
    layers.Dense(10, activation='softmax')
])

model_opt.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# 3. Train with Augmented Data
# Use the training data
history = model_opt.fit(datagen.flow(train_images, train_labels, batch_size=32),
                    epochs=15,
                    validation_data=(test_images, test_labels))

# Save the optimized model
model_opt.save('optimized_model.h5')
print("Model optimized and saved as optimized_model.h5")
```
