```python
# Import necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. NumPy: Numerical Computing
arr = np.array([10, 20, 30, 40, 50])
print(f"Sum of array: {arr.sum()}")
print(f"Minimum value of array: {arr.min()}")
print(f"Max value of array: {arr.max()}")
print(f"Mean of array: {np.mean(arr)}")
print(f"Median of array: {np.median(arr)}")

# 2. Pandas: Data Manipulation
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 78, 92],
    'Hours_Studied': [5, 6, 4, 7]
}
df = pd.DataFrame(data)
print("\nDataset Summary:\n", df.describe())

# 3. Visualization: Matplotlib & Seaborn
plt.figure(figsize=(8, 5))
sns.barplot(x='Name', y='Score', data=df, palette='viridis')
plt.title('Student Scores Comparison')
plt.show()
```
