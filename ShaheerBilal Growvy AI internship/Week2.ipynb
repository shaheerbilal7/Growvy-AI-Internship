```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import mean_squared_error

# Sample Data: Hours studied vs. Test Score
X = np.array([[1], [2], [3], [4], [5], [6]])
y = np.array([35, 42, 50, 58, 70, 82])

# Supervised learning
# Regression
# Linear Regression
lin_reg = LinearRegression()
lin_reg.fit(X, y)

# Decision Tree Regressor
dt_reg = DecisionTreeRegressor()
dt_reg.fit(X, y)

print(f"Linear Reg Prediction: {lin_reg.predict([[7]])[0]:.2f}")
print(f"Decision Tree Prediction: {dt_reg.predict([[7]])[0]:.2f}")

from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Classification
# Load Iris Dataset
iris = load_iris()
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.2, random_state=42)

# Random Forest Classifier
rf_clf = RandomForestClassifier(n_estimators=100)
rf_clf.fit(X_train, y_train)

# Support Vector Machine (SVM)
svm_clf = SVC(kernel='linear')
svm_clf.fit(X_train, y_train)

print(f"Random Forest Accuracy: {accuracy_score(y_test, rf_clf.predict(X_test))*100:.2f}%")
print(f"SVM Accuracy: {accuracy_score(y_test, svm_clf.predict(X_test))*100:.2f}%")

# Unsupervised learning
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

# Using the same Iris data (ignoring labels for unsupervised learning)
kmeans = KMeans(n_clusters=3, random_state=42, n_init='auto')
clusters = kmeans.fit_predict(iris.data)

plt.scatter(iris.data[:, 0], iris.data[:, 1], c=clusters, cmap='viridis')
plt.title("K-Means Clustering (Iris Data)")
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.show()

from sklearn.decomposition import PCA

pca = PCA(n_components=2)
X_pca = pca.fit_transform(iris.data)

print(f"Original shape: {iris.data.shape}")
print(f"Reduced shape (PCA): {X_pca.shape}")
```
